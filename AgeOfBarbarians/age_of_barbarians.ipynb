```python
import os
import pandas as pd
from clickhouse_driver import Client

ck_host = os.environ.get('CK_HOST') or '127.0.0.1'

client = Client(host=ck_host, port=9000, user='wuhao', password='123456')
```

```python
sql = """select count() as n from test.age_of_barbarians;"""
client.query_dataframe(sql)
```

```python
sql = """select round(sumIf(1, pay_count > 0)*100 / uniqExact(user_id),2) pu
from test.age_of_barbarians;"""
client.query_dataframe(sql)
```

```python
sql="""select DATE(register_time) reg_date,count()  n
from test.age_of_barbarians
group by reg_date
order by reg_date"""
client.query_dataframe(sql)
```

```python
sql="""select HOUR(register_time) reg_date,count()  n
from test.age_of_barbarians
group by reg_date
order by reg_date"""
client.query_dataframe(sql)
```

```python
sql="""select round(avgIf(avg_online_minutes,pay_count>0),2) `付费用户在线时长`,
       round(avgIf(avg_online_minutes,pay_count=0),2) `非付费用户在线时长`,
       round(avg(avg_online_minutes),2) `平均在线时长`
from test.age_of_barbarians
"""
client.query_dataframe(sql)
```

```python
sql="""select count() as apa
from test.age_of_barbarians
where pay_count > 0
  and avg_online_minutes > 0;
"""
client.query_dataframe(sql)
```

```python
sql="""select sum(pay_price) / sumIf(1, avg_online_minutes > 0) arppu
from test.age_of_barbarians;
"""
client.query_dataframe(sql)
```

```python
sql="""select sum(pay_price) / sumIf(1, pay_count > 0) arppu
from test.age_of_barbarians;
"""
client.query_dataframe(sql)
```

```python
sql="""select sumIf(1, pay_count > 0) *100 / sumIf(1, avg_online_minutes > 0) pur
from test.age_of_barbarians;
"""
client.query_dataframe(sql)
```

```python
sql="""select count()                       pu,
       sum(pay_price)                sum_pay_price,
       avg(pay_price)                avg_pay_price,
       sum(pay_count)                sum_pay_count,
       avg(pay_count)                avg_pay_count,
       sum_pay_price / sum_pay_count each_pay_price
from test.age_of_barbarians
where pay_price > 0
"""
client.query_dataframe(sql)
```
